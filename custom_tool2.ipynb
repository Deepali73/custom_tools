```python
import os
from langchain.agents import tool, initialize_agent, AgentType
from langchain_google_genai import ChatGoogleGenerativeAI

# Step 1: Set your Gemini API key
os.environ["GOOGLE_API_KEY"] = "AIzaSyCVp"  # Replace with your key

# Step 2: Initialize Gemini Model
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.5)

# Step 3: Create a Custom Recipe Tool
@tool
def get_indian_recipe(dish_name: str) -> str:
    """
    Returns the full recipe for an Indian dish.
    Input should be like: 'sambhar', 'dal makhani', 'poha', etc.
    """
    prompt = (
        f"You're an expert Indian chef. Please provide a detailed recipe for '{dish_name}'. "
        "Include:\n1. List of ingredients\n2. Step-by-step preparation method\n3. Tips for perfect taste.\n"
        "Make it easy to follow for a home cook."
    )
    try:
        return llm.invoke(prompt)
    except Exception as e:
        return f" Error fetching recipe: {str(e)}"

# Step 4: Register the tool
tools = [get_indian_recipe]

# Step 5: Initialize the LangChain Agent
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

# Step 6: Example interaction
if __name__ == "__main__":
    print("Smart Rasoi - Indian Recipe Assistant is ready!\n")
    user_query = input("Ask for a dish (e.g., 'Give me the recipe for sambhar'): ")
    response = agent.run(user_query)
    print(f"\n Recipe:\n{response}")
```
